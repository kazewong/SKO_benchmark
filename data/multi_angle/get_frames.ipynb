```python
import cv2
import os
import matplotlib.pyplot as plt

directory = "/workspaces/highjump-general/highjump_benchmark/data/multi_angle/trimmed"

# List all mp4 files in the directory
mp4_files = [f for f in os.listdir(directory) if f.endswith('.mp4')]

# Specify the frame number to extract (e.g., frame 100)
frame_number = 100
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Extract the specific frame from each mp4
for idx, mp4_file in enumerate(mp4_files):
    video_path = os.path.join(directory, mp4_file)
    cap = cv2.VideoCapture(video_path)
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
    ret, frame = cap.read()
    if ret:
        # Convert BGR (OpenCV) to RGB (matplotlib)
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        axes[idx].imshow(frame_rgb)
        axes[idx].set_title(mp4_file)
        axes[idx].axis('off')
    cap.release()

plt.tight_layout()
plt.show()
```
